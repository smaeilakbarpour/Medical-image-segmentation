```python

def SegmentaionEvaluation1(O_GT, GT, csv_save, save_FIG):
    maskA = GT[:, :, 1]
    maskA[maskA < 200] = 0
    maskA[maskA > 200] = 1
    maskA = np.logical_not(maskA)
    maskM = np.logical_and(GT[:, :, 0], maskA)
    maskN = np.logical_and(GT[:, :, 2], maskA)
        
    maskA_O = O_GT[:, :, 1]
    maskA_O[maskA_O < 200] = 0
    maskA_O[maskA_O > 200] = 1
    maskA_O = np.logical_not(maskA_O)
    # maskM_O = np.logical_and(O_GT[:, :, 0], maskA_O)
    # maskN_O = np.logical_and(O_GT[:, :, 2], maskA_O)
    maskM_O = np.logical_and(O_GT[:, :, 2], maskA_O)  # because maskA_O became logical_not
    maskN_O = np.logical_and(O_GT[:, :, 0], maskA_O)
    
    
    # Calculate evaluation metrics for maskM and maskN
    %run MeasurmentsEvaulation.ipynb
    sensetivity_M, precision_M, specificity_M, accuracy_M, JaccardIndex_M, DiceIndex_M, Fscore_M = MeasurmentsEvaulation(maskM, maskM_O)
    sensetivity_N, precision_N, specificity_N, accuracy_N, JaccardIndex_N, DiceIndex_N, Fscore_N = MeasurmentsEvaulation(maskN, maskN_O)
    
    # Define the parameter names
    paramNames = ['Mask', 'Accuracy', 'Precision', 'Sensitivity', 'Dice Index', 'Jaccard Index', 'Fscore']

    # Create a table with the parameter names and their corresponding values
    resultTable = pd.DataFrame(data=[['maskM', accuracy_M, precision_M, sensetivity_M, DiceIndex_M, JaccardIndex_M, Fscore_M],
                                     ['maskN', accuracy_N, precision_N, sensetivity_N, DiceIndex_N, JaccardIndex_N, Fscore_N]],
                               columns=paramNames)

    # Save the table as a CSV file
    #csv_save = 'C:/Users/NeginRayaneh/Desktop/select_HP_dataset/out_put/SegmentaionEvaluation1_INS_FIG/SegPerformanceINS.csv'
    resultTable.to_csv(csv_save, index=False)
    
    # Print the evaluation metrics for each mask
    for i in range(len(resultTable)):
        print(f"Results for {resultTable['Mask'][i]}:")
        print(f"Accuracy: {resultTable['Accuracy'][i]}")
        print(f"Precision: {resultTable['Precision'][i]}")
        print(f"Sensitivity: {resultTable['Sensitivity'][i]}")
        print(f"Dice Index: {resultTable['Dice Index'][i]}")
        print(f"Jaccard Index: {resultTable['Jaccard Index'][i]}")
        print(f"Fscore: {resultTable['Fscore'][i]}")
        print()
                
    
    maskA_np = np.uint8(maskA * 255)  # Convert binary image to grayscale with pixel values ranging from 0 to 255
    cv2.imwrite(save_FIG+ "maskA(1)_MaskA.png", maskA_np)
    
    maskM = np.uint8(maskM * 255)
    cv2.imwrite(save_FIG + "maskM.png", maskM)

    maskN = np.uint8(maskN * 255)
    cv2.imwrite(save_FIG +"maskN.png", maskN)

        
    maskM_O_astype = (maskM_O * 255).astype(np.uint8)
    maskN_O_astype = (maskN_O * 255).astype(np.uint8)
    maskA_O_astype = (maskA_O * 255).astype(np.uint8)
    cv2.imwrite(save_FIG + "pred_maskM.png", maskM_O_astype)
    cv2.imwrite(save_FIG + "pred_maskN.png", maskN_O_astype)
    cv2.imwrite(save_FIG + "pred_maskA.png", maskA_O_astype)
    

    plt.figure()
    plt.imshow(maskA, cmap='gray')
    plt.title('maskA')
    plt.show()
    
    plt.figure()
    plt.imshow(maskM, cmap='gray')
    plt.title('maskM')
    plt.show()
    
    plt.figure()
    plt.imshow(maskN, cmap='gray')
    plt.title('maskN')
    plt.show()
            
    plt.figure()
    plt.imshow(maskA_O, cmap='gray')
    plt.title('maskA_pred')
    plt.show()
    
    plt.figure()
    plt.imshow(maskM_O, cmap='gray')
    plt.title('maskM_pred')
    plt.show()
    
    plt.figure()
    plt.imshow(maskN_O, cmap='gray')
    plt.title('maskN_pred')
    plt.show()
        
    return        
```
